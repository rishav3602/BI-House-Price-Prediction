```python
## we are mapping all the categorical features 
```

```python
df['Furnishing'].unique()
```

```python
Furnishing_map={"Unfurnished":1,"Semi-Furnished":2,"Furnished":3}
```

```python

df['Status'].unique()
```

```python
Status_map={"Almost_ready":1,"Ready_to_move":2}
```

```python
df['Transaction'].unique()
```

```python
Transaction_map= {"Resale":1,"New_Property":2}
```

```python
df['Furnishing']=df['Furnishing'].map(Furnishing_map)
df['Status'] = df['Status'].map(Status_map)
df['Transaction'] = df['Transaction'].map(Transaction_map)
```
